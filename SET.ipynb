```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
sns.set(style='dark', palette='dark', font="sans-serif", font_scale=1)
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```
