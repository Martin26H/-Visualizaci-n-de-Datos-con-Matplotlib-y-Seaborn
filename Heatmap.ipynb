```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
            vmin=0, vmax=1,cbar=True)
plt.show()
```
