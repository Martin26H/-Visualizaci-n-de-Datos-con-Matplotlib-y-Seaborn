```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax1.set_title('Relacion X-Y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(y,x, 'r')
ax2.set_title('Relacion Y-X')
ax2.set_xlabel('y')
ax2.set_ylabel('x')
plt.show()

fig.tight_layout()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(5,5))
ax1.plot(x,y, 'b',label='sin(x)')
ax1.set_title('Relacion X-Y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax2.plot(y,x, 'r')
ax2.set_title('Relacion Y-X')
ax2.set_xlabel('y')
ax2.set_ylabel('x')
plt.show()

fig.tight_layout()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(5,5))
ax1.plot(x,y, 'b',label='$sin(x)$')
ax1.set_title('Relacion X-Y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax2.plot(y,x, 'r')
ax2.set_title('Relacion Y-X')
ax2.set_xlabel('y')
ax2.set_ylabel('x')
plt.show()

fig.tight_layout()
```
