```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y, 'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y, 'b')
axes[1].plot(y,x, 'r')
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,y)
axes[0,1].plot(y,x, 'r')
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.sin(x),'b')
axes[0,1].plot(x,np.cos(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2,'gray')

#mejorar la visualización de las figuras
fig.tight_layout()
```
